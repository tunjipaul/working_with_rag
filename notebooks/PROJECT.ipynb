## üõ†Ô∏è **WHAT YOU'LL BUILD**

An intelligent AI assistant with **THREE components:**

### 1. AI ENGINE (Agentic RAG with LangGraph)
- Loads and processes policy documents
- Creates a vector database for semantic search
- Builds an agent that decides WHEN to retrieve documents
- Maintains conversation context (follow-up questions work!)
- Cites sources in every answer

### 2. BACKEND (FastAPI)
- RESTful API endpoints for the frontend
- Session management for conversations
- Connects the AI engine to the web

### 3. FRONTEND (React + Tailwind CSS)
- Clean, professional chat interface
- Shows conversation history
- Displays source citations
- Loading states and error handling

## ‚ú® THE MAGIC: Agentic Behavior

Your assistant won't blindly retrieve documents for every question.

It will **THINK:**
- "Is this a greeting? Don't retrieve."
- "Is this asking for policy details? Retrieve documents first."
- "Is this a follow-up question? Use conversation memory."

Just like a smart human assistant would!
